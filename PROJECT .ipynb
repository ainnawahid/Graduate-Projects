### Question 13 
Identify the factors that contribute to neonatal mortality, infant mortality, and maternal mortality rates based on demographic variables?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the dataset containing demographic variables and mortality rates
df = pd.read_excel('cleaned_demographic.xlsx')

# Select the relevant features (demographic variables) and the target variables (mortality rates)
features = ['Crude Birth Rate', 'Crude Death Rate']
target_neonatal = 'Neonatal Mortality Rate'
target_infant = 'Infant Mortality Rate'
target_maternal = 'Maternal Mortality Rate'

# Split the data into training and testing sets
X_train, X_test, y_train_neonatal, y_test_neonatal = train_test_split(df[features], df[target_neonatal], test_size=0.2, random_state=42)
X_train, X_test, y_train_infant, y_test_infant = train_test_split(df[features], df[target_infant], test_size=0.2, random_state=42)
X_train, X_test, y_train_maternal, y_test_maternal = train_test_split(df[features], df[target_maternal], test_size=0.2, random_state=42)

# Create a linear regression model for neonatal mortality rate prediction
model_neonatal = LinearRegression()
model_neonatal.fit(X_train, y_train_neonatal)

# Create a linear regression model for infant mortality rate prediction
model_infant = LinearRegression()
model_infant.fit(X_train, y_train_infant)

# Create a linear regression model for maternal mortality rate prediction
model_maternal = LinearRegression()
model_maternal.fit(X_train, y_train_maternal)

# Predict mortality rates for the test set
y_pred_neonatal = model_neonatal.predict(X_test)
y_pred_infant = model_infant.predict(X_test)
y_pred_maternal = model_maternal.predict(X_test)

# Calculate mean squared error (MSE)
neonatal_mse = np.mean((y_pred_neonatal - y_test_neonatal)**2)
infant_mse = np.mean((y_pred_infant - y_test_infant)**2)
maternal_mse = np.mean((y_pred_maternal - y_test_maternal)**2)

# Plot scatter plots of predicted and actual values with trend lines
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(y_test_neonatal, y_pred_neonatal)
plt.plot([min(y_test_neonatal), max(y_test_neonatal)], [min(y_test_neonatal), max(y_test_neonatal)], color='red', linestyle='--')
plt.xlabel('Actual Neonatal Mortality Rate')
plt.ylabel('Predicted Neonatal Mortality Rate')
plt.title('Neonatal Mortality Rate')

plt.subplot(1, 3, 2)
plt.scatter(y_test_infant, y_pred_infant)
plt.plot([min(y_test_infant), max(y_test_infant)], [min(y_test_infant), max(y_test_infant)], color='red', linestyle='--')
plt.xlabel('Actual Infant Mortality Rate')
plt.ylabel('Predicted Infant Mortality Rate')
plt.title('Infant Mortality Rate')

plt.subplot(1, 3, 3)
plt.scatter(y_test_maternal, y_pred_maternal)
plt.plot([min(y_test_maternal), max(y_test_maternal)], [min(y_test_maternal), max(y_test_maternal)], color='red', linestyle='--')
plt.xlabel('Actual Maternal Mortality Rate')
plt.ylabel('Predicted Maternal Mortality Rate')
plt.title('Maternal Mortality Rate')

plt.tight_layout()
plt.show()

# Print the evaluation results
print("Neonatal Mortality Rate MSE:", neonatal_mse)
print("Infant Mortality Rate MSE:", infant_mse)
print("Maternal Mortality Rate MSE:", maternal_mse)
```

### Question 15
How well does the Random Forest Regression model predict the Total Fertility Rate based on demographic variables?

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset containing demographic variables and Total Fertility Rate
df = pd.read_excel('cleaned_demographic.xlsx')

# Select the relevant features (demographic variables) and the target variable (Total Fertility Rate)
features = ['Crude Rate of Natural Increase', 'Crude Birth Rate', 'Crude Death Rate', 'Neonatal Mortality Rate', 'Infant Mortality Rate', 'Toddler Mortality Rate', 'Maternal Mortality Rate']
target = 'Total Fertility Rate'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Create a Random Forest Regression model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict the Total Fertility Rate for the test set
y_pred = model.predict(X_test)

# Visualize the predicted values against the actual values
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Perfect Predictions')
plt.xlabel('Actual Total Fertility Rate')
plt.ylabel('Predicted Total Fertility Rate')
plt.title('Total Fertility Rate Prediction')
plt.legend()
plt.grid(True)

# Format the y-axis ticks to display decimal numbers
plt.gca().get_yaxis().set_major_formatter(plt.FormatStrFormatter('%.2f'))

# Display the coefficient of determination (R-squared) on the graph
r2 = r2_score(y_test, y_pred)
plt.text(y_test.min(), y_test.max(), f'R-squared = {r2:.2f}', ha='left', va='center', fontsize=12)

plt.show()

# Print the evaluation results
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

- In this analysis, a Random Forest Regression model was developed to predict the Total Fertility Rate using demographic variables. The model's performance was evaluated using two metrics: Mean Squared Error (MSE) and R-squared (R2).

- The MSE value obtained was 0.009238349436364523, indicating a relatively small average squared difference between the predicted and actual Total Fertility Rates. A lower MSE suggests better performance, with a perfect fit having an MSE of 0.

- The R2 value achieved was 0.9953535244500287, indicating that approximately 99.54% of the variance in the Total Fertility Rate can be explained by the demographic variables. A higher R2 value suggests a better fit of the model to the data, with 1 representing a perfect fit.

- These evaluation results highlight the accuracy of the Random Forest Regression model in predicting the Total Fertility Rate based on the provided demographic variables. The low MSE and high R2 value demonstrate the model's effectiveness in capturing the relationship between demographics and fertility rates.

### Question 16 
How well does the linear regression model predict the crude birth rate using demographic variables?

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset containing demographic variables and crude birth rate
df = pd.read_excel('cleaned_demographic.xlsx')

# Select the relevant features (demographic variables) and the target variable (crude birth rate)
features = ['Crude Rate of Natural Increase', 'Crude Death Rate', 'Total Fertility Rate', 'Neonatal Mortality Rate', 'Infant Mortality Rate', 'Toddler Mortality Rate', 'Maternal Mortality Rate']
target = 'Crude Birth Rate'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the crude birth rate for the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print evaluation results
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)

# Visualize the predicted values against the actual values
plt.figure(figsize=(8, 6))
plt.plot(y_test.values, label='Actual', marker='o')
plt.plot(y_pred, label='Predicted', marker='o')
plt.xlabel('Observation')
plt.ylabel('Crude Birth Rate')
plt.title('Crude Birth Rate Prediction')
plt.legend()
plt.grid(True)
plt.show()
```

- A linear regression model was developed to predict the crude birth rate using demographic variables. The model was trained and evaluated using an 80/20 split of the dataset. The evaluation metrics show excellent performance: MSE is 0.0032 and R2 is 0.9999. These results indicate a close alignment between predicted and actual values, with approximately 99.99% of the crude birth rate variance explained by the demographic variables. This demonstrates the strong relationship between demographics and birth rate, highlighting the model's potential for understanding and forecasting population dynamics.
